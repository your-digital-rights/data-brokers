```python
master_list_df['Email'].isnull().sum()
```

```python
merged = pd.merge(master_list_df, full, on=['Domain'], how='left')
merged
```

```python
merged['Email'] = merged['Email_x'].where(merged['Email_x'].notnull(), merged['Email_y'])
merged['Name'] = merged['Name_x'].where(merged['Name_x'].notnull(), merged['Name_y'])
merged['Privacy Policy'] = merged['Privacy Policy_x'].where(merged['Privacy Policy_x'].notnull(), merged['Privacy Policy_y'])
merged = merged.drop(['Privacy Policy_x','Privacy Policy_y','Email_x','Email_y','Name_x','Name_y'], axis=1)
merged
```
