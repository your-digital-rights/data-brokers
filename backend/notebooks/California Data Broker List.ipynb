```python
import pandas as pd
from io import StringIO
import requests


#CALIFORNIA_LIST_URL = "https://www.oag.ca.gov/data-brokers/export?download=1&eid=4362&token=xMURLGqtwqmb0B2gcssOaxYU6uIAnNyegsydSeusXlg"
CALIFORNIA_LIST_PATH = "../data/data-brokers-california.csv"

#request = requests.get(CALIFORNIA_LIST_URL)
#data = request.content.decode('utf-8')
#cali_df = pd.read_csv(StringIO(data), index_col=0 , thousands=',')

cali_df = pd.read_csv(CALIFORNIA_LIST_PATH, parse_dates=['Date Added'], thousands=',')
cali_df
```

```python
cali_df['Email'] =  cali_df.apply(lambda row: row['Email Address'].replace(" [at] ", "@") , axis = 1) 
cali_df
```

```python
from tld import get_tld

def get_domain(row):
    try:
        return get_tld(row['Website URL'], as_object=True).fld
    except Exception as e:
        print(row['Website URL'])

cali_df['Domain'] =  cali_df.apply(get_domain, axis = 1) 
cali_df
```

```python
cali_df.drop(['Email Address','Website URL','How a consumer may opt out of sale or submit requests under the CCPA','Date Added','Additional information about data collecting practices'], axis=1, inplace=True)
cali_df
```

```python
cali_df.rename(columns={'Physical Address': 'Address', 'How a protected individual can demand deletion of information posted online under Gov. Code sections 6208.1(b) or 6254.21(c)(1)': 'Data Broker Comments','Data Broker Name':'Name'}, inplace=True)
cali_df
```

```python
r1 = requests.get('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ2WVw7ygKfibGalx_iDJpsUrd4JnuxycwG61aSXC3SQx4IryJOECbIIXZdPDKqEPQ_MNFyGeFEOgQZ/pub?gid=721859504&single=true&output=csv')
data1 = r1.content.decode('utf-8')
master_list_df = pd.read_csv(StringIO(data1), parse_dates=[0], thousands=',', delimiter=',', header=0, encoding='ascii')#, index_col=0)
master_list_df
```

```python
new_entries = cali_df[~cali_df['Domain'].isin(master_list_df['Domain'])]
new_entries
```

```python
new_entries = new_entries[['Domain', 'Email','Name','Address','Data Broker Comments']]
new_entries.insert(loc=3, column='Categories', value='')
new_entries.insert(loc=4, column='Privacy Policy', value='')
new_entries.insert(loc=5, column='Opt-Out Method', value='')
new_entries.insert(loc=6, column='Opt Out URL', value='')
new_entries.insert(loc=7, column='Phone Number', value='')
new_entries
```
