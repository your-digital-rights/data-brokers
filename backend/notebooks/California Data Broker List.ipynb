```python
import pandas as pd
from io import StringIO
import requests


#CALIFORNIA_LIST_URL = "https://www.oag.ca.gov/data-brokers/export?download=1&eid=4362&token=xMURLGqtwqmb0B2gcssOaxYU6uIAnNyegsydSeusXlg"
CALIFORNIA_LIST_PATH = "../data/data-brokers-california.csv"

#request = requests.get(CALIFORNIA_LIST_URL)
#data = request.content.decode('utf-8')
#cali_df = pd.read_csv(StringIO(data), index_col=0 , thousands=',')

cali_df = pd.read_csv(CALIFORNIA_LIST_PATH, parse_dates=['Date Added'], thousands=',')
cali_df
```

```python
cali_df['Email'] =  cali_df.apply(lambda row: row['Email Address (Accessible to the public)'].replace(" [at] ", "@") , axis = 1) 
cali_df
```

```python
cali_df[cali_df['Website URL'].str.contains("-")]
```

```python
from tld import get_tld

def get_domain(row):
    try:
        return get_tld(row['Website URL'], as_object=True).fld
    except Exception as e:
        print(row['Website URL'])

cali_df['Domain'] =  cali_df.apply(get_domain, axis = 1) 
cali_df
```

```python
cali_df.drop(['Email Address (Accessible to the public)','Website URL','How a consumer may opt out of sale or submit requests under the CCPA','Date Added','Additional information about data collecting practices'], axis=1, inplace=True)
cali_df
```

```python
cali_df.rename(columns={'Physical Address': 'Address', 'How a protected individual can demand deletion of information posted online under Gov. Code sections 6208.1(b) or 6254.21(c)(1)': 'DB Comments','Data Broker Name':'Name'}, inplace=True)
cali_df
```

```python
r1 = requests.get('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ2WVw7ygKfibGalx_iDJpsUrd4JnuxycwG61aSXC3SQx4IryJOECbIIXZdPDKqEPQ_MNFyGeFEOgQZ/pub?gid=721859504&single=true&output=csv')
data1 = r1.content.decode('utf-8')
master_list_df = pd.read_csv(StringIO(data1), parse_dates=[0], thousands=',', delimiter=',', header=0, encoding='ascii')#, index_col=0)
master_list_df
```

```python
new_entries = cali_df[~cali_df['Domain'].isin(master_list_df['Domain'])]
new_entries
```

```python
master_list_df['Email'].isnull().sum()
```

```python
merged = pd.merge(master_list_df, cali_df, on=['Domain'], how='left')
merged
```

```python
merged['Email'] = merged['Email_x'].where(merged['Email_x'].notnull(), merged['Email_y'])
merged['Name'] = merged['Name_x'].where(merged['Name_x'].notnull(), merged['Name_y'])
merged['Address'] = merged['Address_x'].where(merged['Address_x'].notnull(), merged['Address_y'])
merged = merged.drop(['Address_x','Address_y','Email_x','Email_y','Name_x','Name_y'], axis=1)
merged
```

```python
merged = merged.append(new_entries, ignore_index=True)
merged
```

```python
merged.drop_duplicates(subset='Domain', inplace=True)
merged
```

```python
merged.to_csv('data-brokers-merged.csv')
```
